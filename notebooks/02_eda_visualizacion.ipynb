#### Gráfico de conteo por año

##### Convertimos las columnas que contienen fechas en tipo date

```python
df["FE_PRESEN_RECLA"] = pd.to_datetime(df["FE_PRESEN_RECLA"], format="%m/%d/%Y")
df["FE_RESULT_RECL"] = pd.to_datetime(df["FE_RESULT_RECL"], format="%m/%d/%Y")
```

```python
df["FE_NOTIFI_RESULT"] = df["FE_NOTIFI_RESULT"].astype(str).str.strip().str.replace('"', '')
```

```python
df["FE_NOTIFI_RESULT"] = pd.to_datetime(df["FE_NOTIFI_RESULT"], format="%Y%m%d")
```

##### Extramoes el año para la visualización

```python
df["FE_PRESEN_RECLA"].dtype
```

```python
df["AÑO"] = df["FE_PRESEN_RECLA"].dt.year
```

```python
df_filtered = df[df["AÑO"].isin([2021, 2022, 2023, 2024])]
```

#### Distribucion de reclamos por año

```python
sns.countplot(
    data=df_filtered, 
    x="AÑO",
    hue="AÑO",
    legend=False
)
plt.title("Distribución de Reclamos por Año")
plt.xlabel("Año")
plt.ylabel("Cantidad de Reclamos")
```

#### Distribucion de reclamos por mes

```python
df["MES"] = df["FE_PRESEN_RECLA"].dt.month
```

```python
#df_filtered["MES"] = df["FE_PRESEN_RECLA"].dt.month
#df_filtered.loc[:, "MES"] = df["FE_PRESEN_RECLA"].dt.month

sns.countplot(
    data=df_filtered, 
    x="MES",
    hue="MES",
    legend=False
)
plt.title("Distribución de Reclamos por Mes")
plt.xlabel("Mes")
plt.ylabel("Cantidad de Reclamos")
plt.xticks(ticks=range(12), labels=["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])

plt.show()
```

#### Tendencia de reclamos por mes desde el 2021

```python
#df_filtered["AÑO_MES"] = df_filtered["AÑO"].astype(str) + "-" + df_filtered["MES"].astype(str)
df_filtered = df[df["AÑO"] >= 2021].copy()
df_filtered.loc[:, "AÑO_MES"] = df_filtered["AÑO"].astype(str) + "-" + df_filtered["MES"].astype(str)
reclamos_por_mes = df_filtered.groupby("AÑO_MES").size()

plt.figure(figsize=(12,5))
sns.lineplot(x=reclamos_por_mes.index, y=reclamos_por_mes.values, marker="o")

plt.xticks(rotation=45)
plt.title("Tendencia de Reclamos por Mes")
plt.xlabel("Año-Mes")
plt.ylabel("Cantidad de Reclamos")
plt.show()
```

#### Medio de presentación de Reclamos

```python
sns.countplot(data=df_filtered, y="DE_MEDIO_PRESENTACION", 
              hue="DE_MEDIO_PRESENTACION", 
              order=df_filtered["DE_MEDIO_PRESENTACION"].value_counts().index, 
              palette="coolwarm",
              legend=False)

plt.title("Medio de Presentación de Reclamos")
plt.xlabel("Cantidad de Reclamos")
plt.ylabel("Medio de Presentación")
plt.show()
```

#### Estado de los reclamos

```python
sns.countplot(data=df_filtered, x="DE_ESTADO_RECLAMO",
              hue="DE_ESTADO_RECLAMO",
              palette="Set2",
              legend=False)

plt.title("Estado de los Reclamos")
plt.xlabel("Estado")
plt.ylabel("Cantidad de Reclamos")
plt.xticks(rotation=45)
plt.show()
```

#### Duracion de los reclamos segun su estado

```python
# Calcular la duración del reclamo en días
df_filtered["DURACION_RECLAMO"] = (df_filtered["FE_RESULT_RECL"] - df_filtered["FE_PRESEN_RECLA"]).dt.days
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_filtered,
            hue="DE_ESTADO_RECLAMO",
            x="DE_ESTADO_RECLAMO", 
            y="DURACION_RECLAMO", 
            palette="Set3",
           legend=False)

plt.title("Duración de los Reclamos según su Estado")
plt.xlabel("Estado del Reclamo")
plt.ylabel("Duración (días)")
plt.xticks(rotation=45)
plt.ylim(0, 1000)
plt.show()
```

#### Reclamos por servicio

```python
plt.figure(figsize=(12, 6))
top_servicios = df["DE_SERVICIO"].value_counts().head(10)
sns.barplot(
    x=top_servicios.values, 
    y=top_servicios.index, 
    hue=top_servicios.index,
    legend=False,
    palette="magma")
plt.title("Top 10 servicios con más reclamos")
plt.xlabel("Cantidad de reclamos")
plt.ylabel("Servicio")
plt.show()
```

#### Correlacion de variables

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlaciones entre variables numéricas")
plt.show()
```
